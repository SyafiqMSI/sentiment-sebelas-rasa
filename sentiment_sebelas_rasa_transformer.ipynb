```python
# Load model directly
from transformers import AutoTokenizer, AutoModelForSequenceClassification

tokenizer = AutoTokenizer.from_pretrained("arifagustyawan/sentiment-roberta-id")
model = AutoModelForSequenceClassification.from_pretrained("arifagustyawan/sentiment-roberta-id")
```

```python
#load data
import json

data_path = 'backup/instagram_tagged_posts_20241229_195451.json'

# Open and read the JSON file
with open(data_path, 'r', encoding='utf-8') as file:
    data = json.load(file)

# Print the data
print(data)
```

```python
import re

comments = []
non_replyComments = []

for i in range(len(data)):
  for j in range(len(data[i]['comments'])):
    result = re.sub(r"@\w+", "", data[i]['comments'][j]['comment'])
    cleaned_text = re.sub(r"\s{2,}", " ", result).strip()

    if data[i]['comments'][j]['is_reply'] == False:
      non_replyComments.append(data[i]['comments'][j]['comment'])
    comments.append(data[i]['comments'][j]['comment'])
```

```python
!pip install torch
```

```python
prediction = ['positive', 'neutral', 'negative']
```

```python
import torch

def predict(text):
  # Tokenize the input text
  inputs = tokenizer(text, return_tensors="pt", truncation=True, padding=True, max_length=512)

  # Perform inference
  with torch.no_grad():
      outputs = model(**inputs)
      logits = outputs.logits

  # Convert logits to probabilities (optional)
  probabilities = torch.nn.functional.softmax(logits, dim=-1)

  # Get the predicted class (highest probability)
  predicted_class = torch.argmax(probabilities, dim=-1).item()

  return logits, probabilities, predicted_class
```

```python
# Print results
logits, probabilities, predicted_class = predict(comments[0])

print(comments[0])
print("Logits:", logits)
print("Probabilities:", probabilities)
print("Predicted class:", predicted_class)
print("Prediction", prediction[predicted_class])
```

```python
#predict
comments_result = []
non_replyComments_result = []

for comment in comments:
    logits, probabilities, predicted_class = predict(comment)
    comments_result.append([comment, probabilities[0, 0].item(), probabilities[0, 1].item(), probabilities[0, 2].item(), prediction[predicted_class]])

for comment in non_replyComments:
    logits, probabilities, predicted_class = predict(comment)
    non_replyComments_result.append([comment, probabilities[0, 0].item(), probabilities[0, 1].item(), probabilities[0, 2].item(), prediction[predicted_class]])
```

```python
import pandas as pd

comments_result_df = pd.DataFrame(comments_result, columns=["Comment", "Positive", "Neutral", "Negative", "Sentiment"])
non_replyComments_result_df = pd.DataFrame(non_replyComments_result, columns=["Comment", "Positive", "Neutral", "Negative", "Sentiment"])
```
